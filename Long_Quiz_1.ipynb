```python
class Student:
  def class_standing(recitation, quiz, assignment, seatwork, perofrmance_task, midterm_project, final_project):
    return 0.05 * recitation + 0.05 * quiz + 0.05 * assignment + 0.5 * seatwork + 0.2 * perofrmance_task + 0.3 * midterm_project + 0.3 * final_project

def prelim_grade(prelim_exam, prelim_cs):
    return 0.5 * prelim_exam + 0.5 * prelim_cs

def midterm_grade(prelim, midterm_exam, midterm_cs):
    return (1/3 * prelim) + (2/3 * (0.5 * midterm_exam + 0.5 * midterm_cs))

def final_grade(midterm, final_exam, final_cs):
    return (1/3 * midterm) + (2/3 * (0.5 * final_exam + 0.5 * final_cs))

def convert_to_ucc_grade(final):
    grades = [(98, 1.0), (95, 1.25), (90, 1.5), (85, 1.75), (80, 2.0),
              (75, 2.25), (70, 2.5), (65, 2.75), (60, 3.0)]
    for score, ucc in grades:
        if final >= score:
            return ucc
    return 5.0


def input_and_calculate_grades():
    id = input("Enter the ID of the student: ")
    name = input("Enter the name of the student: ")
    program = input("Enter the program of the student: ")
    course = input("Enter the course of the student: ")
    sex = input("Enter the sex of the student: (M/F) ")
    weight = (input("Enter the weight of the student: (in kg) "))
    height = (input("Enter the height of the student: (in cm) "))
    recitation = float(input("Enter the recitation score: "))
    seatwork = input(("Enter the Seatwork score: "))
    performance_task = float(input("Enter the Performance Task score: "))
    midterm_project = float(input("Enter the Midterm Project score: "))
    final_project = float(input("Enter the Final Project score: "))
    quiz = float(input("Enter the Quiz score: "))
    assignment = float(input("Enter the Assignment score: "))
    prelim_exam = float(input("Enter the Prelim Exam score: "))
    midterm_exam = float(input("Enter the Midterm Exam score: "))
    final_exam = float(input("Enter the Final Exam score: "))

    prelim_cs = class_standing(recitation, quiz, assignment)
    prelim = prelim_grade(prelim_exam, prelim_cs)
    midterm_cs = class_standing(recitation, quiz, assignment)
    midterm = midterm_grade(prelim, midterm_exam, midterm_cs)
    final_cs = class_standing(recitation, quiz, assignment)
    final = final_grade(midterm, final_exam, final_cs)
    ucc = convert_to_ucc_grade(final)


    status = "Passed" if final >= 60 else "Failed"

    print(f"\nStudent: {name}")
    print(f"Prelim Grade: {prelim:.2f}")
    print(f"Midterm Grade: {midterm:.2f}")
    print(f"Final Grade: {final:.2f}")
    print(f"CPE 103: {ucc:.2f}")
    print(f"CPE 103: {status}\n")


print("Student 1:")
input_and_calculate_grades()
print("Student 2:")
input_and_calculate_grades()
print("Student 3:")
input_and_calculate_grades()
```
